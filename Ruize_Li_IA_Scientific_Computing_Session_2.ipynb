```python
import numpy as np
import matplotlib.pyplot as plt

# locate data

exp02=np.genfromtxt('session2data.dat', delimiter=',', skip_header=2)

# Name data indicating time, apple, IBM, microsoft
# only need time from 2000 to the end of 2015, so row No. 10 to 202
t=exp02[10:202,0]
a=exp02[10:202,1]
i=exp02[10:202,2]
m=exp02[10:202,3]
plt.title("Comparison of share prices between the start of 2000 and the end of 2015")
plt.xlabel("Time in years")
plt.ylabel("Share prices in US dollars for given companies")
plt.plot(t,a,'-r',label="Apple")
plt.plot(t,i,'-b',label="IBM")
plt.plot(t,m,'-g',label="Microsoft")
plt.legend(loc='upper left')
plt.savefig('session2fig.png',dpi=300)
plt.show()
```
