```python
import numpy as np
import matplotlib.pyplot as plt

# creating variables x and y
x = np.arange(-10,10,0.5)
y = x**3 + 2*x**2 + 3*x + 4

# Label axes and give graph title
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("The graph of y=y(x) from -10 to 10")

# draw the graph
plt.show()

# the additional plot is shown in the next code window
```

```python
import numpy as np
import matplotlib.pyplot as plt

# creating x, y, f, and g
x = np.arange(-10,10,0.5)
y = x**3 + 2*x**2 + 3*x + 4
f = x**3 + 1*x**2 + 3*x + 4
g = x**3 + 3*x**2 + 3*x + 4

# labeling axes and giving the final three-line graph
plt.plot(x,y,label="quad coefficient = 2")
plt.plot(x,f,label="quad coefficient = 1")
plt.plot(x,g,label="quad coefficient = 3")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("The effect of quadratic coefficient")

plt.legend()

plt.show()
```
